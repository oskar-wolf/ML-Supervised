```python
#load required packages for logistic regression
from sklearn import datasets
import numpy as np
from sklearn.linear_model import LogisticRegression
from matplotlib import pyplot as plt
```

```python
#load data
iris = datasets.load_iris()
iris
```

```python
#prepare the data
X = iris['data'][:,3:]#petal width
y = (iris['target'] == 2).astype(np.int32) #1 if Iris virginica, else 0
```

```python
X
```

```python
y
```

```python
#define and fit model
log_reg = LogisticRegression()
log_reg.fit(X,y)
```

```python
#generate new data to base predictions on 
X_new = np.linspace(0,3,1000).reshape(-1,1)
```

```python
#predict probablilities for new data
y_proba = log_reg.predict_proba(X_new)
```

```python
plt.plot(X_new, y_proba[:,1],'g-',label = 'Iris virginica')
plt.plot(X_new,y_proba[:,0],'b--',label='NotIris virginica')
```
